```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('zomato.csv')
df.head(2)
```

# Data cleaning (Exploratory analysis)

```python
# drop unnecessary column
df.drop(['url', 'address', 'phone', 'dish_liked', 'reviews_list', 'menu_item'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

## **There will be some null value in column let hadle it:**

## **Handle null value of rate**

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

## **replace null value with the mean of rate calumn**

```python
# there are '10019' null value present in rate column
# replace the null value with its mean

df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

## **Drop null value**

```python
df.dropna(inplace=True)
df.info()
```

```python
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plate', 'listed_in(type)':'type'}, inplace=True)
df.head()
```

**there are two column of city so we drop the listed_city column**

```python
df.drop(['listed_in(city)'], axis=1, inplace=True)
df.head()
```

```python
df['cost2plate'].unique()
```

## **Handle comma in cost2plate column**

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost2plate'] = df['cost2plate'].apply(handlecomma)
df['cost2plate'].unique()
```

```python
df.head()
```

## **Clean rest type column**

```python
df['rest_type'].unique()

rest_type = df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_typelessthan1000 = rest_type[rest_type<1000]
rest_typelessthan1000
```

```python
def handle_resttype(value):
    if value in rest_typelessthan1000:
        return 'other'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_resttype)
df['rest_type'].value_counts()
```

## **cleaning location column**

```python
df.head()
```

```python
location = df['location'].value_counts(ascending=False)
```

```python
locationlessthan300 = location[location<300]

def handle_location(value):
    if value in locationlessthan300:
        return 'other'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts(ascending=False)
```

## **Cleaning cuisins column**

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending=False)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'other'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts(ascending=False)
```

```python
# type is clean already
df['type'].value_counts()
```
